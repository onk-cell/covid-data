```python
df = pd.read_csv('../input/covid-data/StatewiseTestingDetails.csv')
df
```

```python
sorted = df.sort_values('Positive', ascending=False)
sorted = sorted[sorted['Date']=='2020-04-17']
states = sorted[['Date', 'State', 'Positive','TotalSamples', 'Negative']]
states.head()
```

```python
states_10 = states.head(10)
```

no of positive cases by state 

```python
import matplotlib.pyplot as plt
import seaborn as sns
f, ax = plt.subplots(figsize=(10, 5))
sns.set_color_codes("pastel")
sns.barplot(x="Positive", y="State", data=states_10,
            label="Positive", color="b")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
f, ax = plt.subplots(figsize=(10, 5))
sns.set_color_codes("pastel")
sns.barplot(x="Negative", y="TotalSamples", data=states_10,
            label="TotalSamples", color="b")
```
